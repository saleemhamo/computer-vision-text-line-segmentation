# ahte_test_diva_ground_image

```python

data_path = '/content/drive/MyDrive/Dataset/ahte_dataset/ahte_test_diva_ground_image'

images1 = []

rootdir = data_path
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        frame = cv.imread(os.path.join(subdir, file),cv.IMREAD_GRAYSCALE) 
        images1.append(frame)


cv2_imshow(images1[0])
print(len(images1))
```

```python


# ret, bw_img = cv.threshold(images1[0], 240, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C)
ret, bw_img = cv.threshold(images1[0], 0, 255, cv.THRESH_BINARY)

cv2_imshow(bw_img)

final_frame = cv.hconcat((images1[0], bw_img))
cv2_imshow(final_frame)
# ret, image1 = cv.threshold(images1[0], 120, 255, cv.THRESH_BINARY)
# # final_frame = cv2.hconcat((thresh1, image))
# cv2_imshow(image1)
```

```python
kernel = np.ones((5,2),np.float32)

edges =  cv.dilate(bw_img,kernel,iterations = 1)
edges = cv.morphologyEx(edges, cv.MORPH_CLOSE, kernel)
edges = cv.morphologyEx(edges, cv.MORPH_CLOSE, kernel)
cv2_imshow(edges)
edges = cv.Canny(edges,100,200)
cv2_imshow(edges)


# kernel = np.ones((7,2),np.float32)/20
# ret2, bw_img2 = cv.threshold(bw_img, 0, 255, cv.THRESH_BINARY_INV)

# # cv2_imshow(bw_img2-edges)
# # cv2_imshow(bw_img2)
# avg_img = cv.filter2D(bw_img2,-1,kernel)
# cv2_imshow(avg_img)

# # out1=avg_img +edges
# # cv2_imshow(out1 )
# # avg_img1 = cv.filter2D(out1,-1,kernel)
# # cv2_imshow(avg_img1)
```
